```python
import seaborn as sb
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fd=pd.read_csv('full_data.csv')
```

```python
fd.head()
```

```python
#Correlation between favorite and retweets
axx=fd[fd['favorites']!=0]
ax = sb.regplot(x="favorites", y="retweets", data=axx)
```

It is obvious that there is a strong correlation between the two values. 
More retweets mean more favorite.

```python
#axx=fd[fd['Dog_style']!='None']

# Draw a violinplot
sb.violinplot(data=fd, x='Dog_style',y='favorites',palette="Set3", bw=.5, cut=1, linewidth=1)

# Finalize the figure
ax.set(ylim=(0, 1.05))
sb.despine(left=True, bottom=True)
```

We can see that 'doggo' and 'puppo' styles collect more 'favorites'

```python
breed = fd.groupby('True_prediction').filter(lambda x: len(x) >= 40)
breed=breed[breed['favorites']<12000]
```

```python
g=sb.boxplot(x='True_prediction', y="favorites",data=breed)
plt.xticks(rotation=45) ;
```

We can see that the most popular breed is the Chow and the most upopular 
is Chihuahua
